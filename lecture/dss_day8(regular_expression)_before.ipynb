* 기본 패턴
 - a, X, 9 기본 character들은 정확히 해당 문자와 일치
 - ^ $ * + [] \ | () 등의 특수문자 제외
 
 - . (마침표) - 어떤 한개의 character와 일치
 
 - \w - 문자 character와 일치 [a-zA-Z0-9_]
 - \s - 공백문자와 일치
 - \t, \n, \r - tab, newline, return
 - \d - 숫자 character와 일치 [0-9]
 - ^ = 시작, $ = 끝 각각 문자열의 시작과 끝을 의미
 - \가 붙으면 스페셜한 의미가 없어짐. 예를들어 \.는 .자체를 의미 \\는 \를 의미

* 반복패턴
 - '+' -> 1번 이상의 패턴이 발생
 - '*' -> 0번 이상의 패턴이 발생
 - ? -> 0 혹은 1번의 패턴이 발생

* search method
 - 첫번째로 패턴을 찾으면 멈춤

```python
import re

match = re.search(r'iii', 'piiig') # =====> 찾고자 하는 문자열 앞 'r'
print match
print match.group() # =====> 표현할떄 .group() 함수 이용

match = re.search(r'iiiig', 'piiig')
print match
```

```python
m = re.search(r'..g', 'piiig') # ======> .은 한개의 문자와 일치
print m.group()

m = re.search(r'\d\d\d', 'p123g')  # =====> \d는 숫자를 표현
print m.group()

m = re.search(r'\d\d\d', '오마이갓123이럴수가') 
print m.group()

m = re.search(r'\w\w\w', '@@abcd!!') # =====> \w는 문자와 일치
print m.group()

m = re.search(r'\w\w\w', '@@ab0!!')
print m.group()
```

```python
m = re.search(r'pi+', 'piiiiiiiiig')
print m.group()
```

```python
m = re.search(r'i+', 'piiiiiiiigiiii')
print m.group()
```

```python
m = re.search(r'\d\s*\d\s*\d', 'xx1   2   3xx')
print m.group()
```

```python
m = re.search(r'^b\w+', 'banana')
print m.group()

m = re.search(r'b\w+', 'foobar')
print m.group()
```

* 연습문제) 
 - 기자 이메일 주소 뽑아오기
 - 전화번호 패턴의 정규표현식을 만들어봅시다.

```python
# pip install requests
# pip install bs4 

import requests
from bs4 import BeautifulSoup

def get_news_content(url):
    response = requests.get(url)
    content = response.text

    soup = BeautifulSoup(content)

    div = soup.find('div', attrs = {'id' : 'harmonyContainer'})
    
    content = ''
    for paragraph in div.find_all('p'):
        content += paragraph.get_text()
        
    return content.encode('utf-8')
        
news1 = get_news_content('http://v.media.daum.net/v/20170112152314639')
print news1
```

* findall
 - search가 최초로 매칭되는 패턴만 반환한다면, findall은 매칭되는 전체의 패턴을 반환한다.
 - 리스트 형태로 반환

```python
str = 'What a nice weather macmath22@gmail.com, test@test.com mina@minas.net'
emails = re.findall(r'[\w\.-]+@[\w\.-]+', str) 
print emails
for email in emails:
    print email
```

```python
f = open('./data_set/tf_idf1.txt', 'r')

content = f.read()
strings = re.findall(r'[p]ython', content)
print strings

strings = re.findall(r'[P]ython', content)
print strings

strings = re.findall(r'[pP]ython', content)
print strings

strings = re.findall(r'python', content, re.IGNORECASE)
print strings

f.close()
```

```python
str = 'What a nice weather macmath22@gmail.com, test@test.com mina@minas.net'
emails = re.findall(r'([\w\.-]+)@([\w\.-]+)', str) # 그룹( )이 있으면 튜플의 리스트로 반환
print emails                                       # 그룹( )이 없는 그냥 경우에는 리스트로 반환
for email in emails:
    print email, email[0], email[1] 
```

* sub
 - 주어진 문자열에서 일치하는 모든 패턴을 replace
 - 그 결과를 문자열로 다시 반환함
 - 두번째 인자는 특정 문자열이 될 수도 있고, 함수가 될 수 도 있다.
 - count가 0인 경우는 전체를, 1이상이면 해당 숫자만큼 치환된다.
 - re.sub(pattern, repl, string, count=0, flags=0)

```python
str = 'What a nice weather macmath22@gmail.com, test@test.com mina@minas.net'
replaced = re.sub(r'[\w\.-]+@[\w\.-]+', r'test', str) 
print replaced
```

* 실습)
 - 010-3335-5555형식의 번호를 (010) 3335-5555로 변환하는 함수를 만드시오
 - grouping을 이용

```python
pattern = r'(\d{3})-(\d{4})-(\d{4})'
replace = r'(\1) \2-\3' # \1,2,3 \하고 그룹번호 쓰면 된다.

result = re.sub(pattern, replace, '010-3457-6360')
print result
```

* compile 
 - 동일한 정규표현식을 매번 다시 쓰기 번거로움을 해결
 - compile로 해당표현식을 re.RegexObject 객체로 저장하여 사용가능

```python
import re


email_reg = re.compile(r'[\w.-]+@[\w.-]+')
m = email_reg.search('What a nice weather macmath22@gmail.com, test@test.com mina@minas.net')
if m:
    print m.group()
    

print email_reg.findall('What a nice weather macmath22@gmail.com, test@test.com mina@minas.net')

print email_reg.sub('test', 'My email is macmath22@gmail.com')
```
