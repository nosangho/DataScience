#2. numpy 버젼

```python
import numpy as np

steps = 1000
way = np.random.randint(0,2,size = steps) # 각 걸음이 0인지 1인지
step = np.where(way>0 ,1, -1)             # 1이면 +1 , 0이면 -1 array 반환
walk = step.cumsum()                      # step의 누적합(cumsum)

print walk
```
